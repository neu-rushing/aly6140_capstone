```python
from IPython.display import display, HTML, clear_output

# HTML and JavaScript code for the timer
timer_html = """
<div id="timer">Timer: <span id="countdown"></span></div>
<script>
var timeLeft = 300; // 5 minutes
var timer = setInterval(function() {
    document.getElementById("countdown").textContent = timeLeft + 's';
    timeLeft--;
    if (timeLeft < 0) {
        clearInterval(timer);
        alert("Time's up!");
        document.getElementById("answer").disabled = true; // Disable answer input
    }
}, 1000);
</script>
"""

# Display the timer at the top of the input screen
display(HTML(timer_html))
```

```python
def start_game():
    score = 0
    time_limit = 300  # 5 minutes

    print("Welcome to the Trivia Game!")

    for question in random.sample(questions, 10):
        print(question['question'])
        for i, option in enumerate(question['options'], 1):
            print(f"{i}. {option}")

        user_answer = input("Your answer (1-4): ")

        if user_answer.isdigit() and 1 <= int(user_answer) <= 4:
            if question['options'][int(user_answer) - 1] == question['answer']:
                print("Correct!")
                score += 1
            else:
                print(f"Wrong! The correct answer is: {question['answer']}")
        else:
            print("Invalid input. Skipping this question.")

        time_limit -= 30  # Deduct 30 seconds for each question

    print(f"Game Over! Your score: {score}")

# Start the game
start_game()
```
