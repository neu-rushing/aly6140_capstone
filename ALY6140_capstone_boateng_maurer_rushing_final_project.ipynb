```python
# Initialize the set to store unique usernames
unique_usernames = set()

def registration():
    
    # Get the username from the user
    username = input("Please choose a unique username: ")

    # Check if the username is already in the set
    if username in unique_usernames:
        print("Duplicate username, please try again")
    else:
        # Add the username to the set if it's unique
        unique_usernames.add(username)
        print("Success! Username is unique.")
```

```python
# Create an empty list to store score data
score_data_list = []

def start_game(unique_usernames, score_data_list):
    score = 0
    username = ""

    while True:
        username = input("Please enter your username: ")
        if username in unique_usernames:
            break  # Continue with the game
        else:
            print("Username not found. The game can only be played by registered users.")
            return

    print("Welcome to the Trivia Game!")

    for question in random.sample(questions, 10):
        print(question['question'])
        for i, option in enumerate(question['options'], 1):
            print(f"{i}. {option}")

        user_answer = input("Your answer (1-4): ")

        if user_answer.isdigit() and 1 <= int(user_answer) <= 4:
            if question['options'][int(user_answer) - 1] == question['answer']:
                print("Correct!")
                score += 1
            else:
                print(f"Wrong! The correct answer is: {question['answer']}")
        else:
            print("Invalid input. Skipping this question.")

    print(f"Game Over! Your score: {score}")

    # Create the score data as a string
    score_data = f"{username};{score}"

    # Append the score data to the list
    score_data_list.append(score_data)
```

```python
def display_leaderboard():
    print("Leaderboard:")
    
    # Sort the score_data_list by score (in descending order)
    sorted_scores = sorted(score_data_list, key=lambda data: int(data.split(";")[1]), reverse=True)
    
    # Display the top 5 scores
    for rank, data in enumerate(sorted_scores[:5], start=1):
        username, score = data.split(";")
        print(f"{rank}. {username}: {score}")
    
    return score_data_list
```

```python
import os

def clear_screen():
    if os.name == 'posix':  # Unix/Linux/MacOS
        os.system('clear')
    elif os.name == 'nt':  # Windows
        os.system('cls')
    else:
        print("\n" * 100)  # Fallback: print newlines to "clear" the screen

while True:
    clear_screen()
    print(logo)
    print("1. Registration")
    print("2. View Leaderboard")
    print("3. Play Trivia")
    print("4. Exit")

    user_choice = input("Enter the number of the option you want to run: ")

    if user_choice == '1':
        registration()
    elif user_choice == '2':
        display_leaderboard()
    elif user_choice == '3':
        start_game(unique_usernames, score_data_list)
    elif user_choice == '4':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a valid option (1, 2, 3, or 4).")

```
